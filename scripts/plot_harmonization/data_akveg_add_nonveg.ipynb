```python
# join AKVEG soil, water, rock data to pft data
# akveg = Site Code
# blm = PlotID
```

```python
import pandas as pd
import numpy as np
```

```python
#import akveg pft
p = '/mnt/poseidon/remotesensing/arctic/data/training/Test_03/'
f = 'AKVEG_pft_fcover_00.csv'
akveg = pd.read_csv(p + f, index_col=0)
akveg
```

```python
p = '/mnt/poseidon/remotesensing/arctic/data/vectors/AKVEG_ACCS/'
f = 'BLM_Natl_AIM_TerrADat_Hub.csv'
blm = pd.read_csv(p + f, index_col=5)
blm
```

```python
blm.columns.to_list()
```

```python
concated = pd.concat([akveg, blm], axis=1, join='inner')
nonveg = concated[['BareSoilCover', 'FH_WaterCover', 'FH_RockCover', 
                   'DateVisited', 'TotalFoliarCover']]
nonveg.columns = ['soil cover (%)', 'water cover (%)', 'rock cover (%)', 
                  'DateVisited', 'total cover (%)']
nonveg = nonveg.copy()
nonveg['Year'] = pd.DatetimeIndex(nonveg['DateVisited']).year
nonveg.drop(columns=['DateVisited'], inplace=True)
```

```python
nonveg
```

```python
final = pd.concat([akveg, nonveg], axis=1)
cover = final.filter(regex=("cover"))
cover.sum(axis=1)
```

```python
final
```

```python
p = '/mnt/poseidon/remotesensing/arctic/data/training/Test_03/'
f = 'AKVEG_pft_fcover_01.csv'
final.index.name = 'Site Code'
final.to_csv(p + f)
```
